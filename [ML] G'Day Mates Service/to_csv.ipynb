```python
import os
import re
import pandas as pd
```

```python
input_dir = "./data/txt_raw" 
output_dir = "./data/txt2csv" 

txt_files = [f for f in os.listdir(input_dir) if f.endswith(".txt")]

for txt_file in txt_files:
    input_path = os.path.join(input_dir, txt_file)
    output_file_name = os.path.splitext(txt_file)[0] + ".csv" 
    output_path = os.path.join(output_dir, output_file_name)
    
    data = []  
    
    with open(input_path, "r", encoding="utf-8") as file:
        for line in file:
            match = re.match(r"(.*?), (.*?) : (.*)", line.strip())
            if match:
                date, user, message = match.groups()
                data.append({"Date": date, "User": user, "Message": message})
    
    if data:  # 데이터가 비어 있지 않을 때만 저장
        df_tmp = pd.DataFrame(data)
        df_tmp.to_csv(output_path, index=False, encoding="utf-8-sig")
        print(f"파일 변환 완료: {output_file_name}")
    else:
        print(f"데이터 없음, 변환 생략: {txt_file}")
```
